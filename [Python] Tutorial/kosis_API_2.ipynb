## 3. 통계자료 조회

### 3-1. 최신 시점 찾기
- 현재 기준으로 가장 최신 통계가 언제의 통계인지 확인한다.
- '횡단면 조회' 자료를 받아 최신 통계 시점을 확인한다.
- 통계청 OPEN API 사이트에서 횡단면 조회 query 예제 복사

```
# 0:apiKey, 1:itmId, 2:objL1, 3:objL2, 4:prdSe, 5:loadGubun, 6:ordId, 7:tblId
query = "https://kosis.kr/openapi/Param/statisticsParameterData.do?method=getList&apiKey={0}&itmId={1}+&objL1={2}+&objL2={3}&objL3=&objL4=&objL5=&objL6=&objL7=&objL8=&format=json&jsonVD=Y&prdSe={4}&newEstPrdCnt=1&loadGubun={5}&orgId={6}&tblId={7}"
```

- 통계청 OPEN API 사이트에서 항목, 분류 value 찾기

```
# 항목
itm_id = {
    "인구":"T10",
    "경제활동인구":"T20",
    "취업자":"T30",
    "실업자":"T40",
    "비경제활동인구":"T50",
    "경제활동참가율":"T60",
    "실업률":"T80",
    "고용률":"T90"
}

# 성별 분류
sex_id = {
  "계": "0",
  "남자": "2",
  "여자": "3"
}

# 연령 분류
age_id = {
     "15세 이상 전체": "00",
     "15 - 19세": "10",
     "20 - 29세": "20",
     "20 - 24세": "201",
     "25 - 29세": "202",
     "30 - 39세": "30",
     "30 - 34세": "301",
     "35 - 39세": "302",
     "40 - 49세": "40",
     "40 - 44세": "401",
     "45 - 49세": "402",
     "50 - 59세": "50",
     "50 - 54세": "501",
     "55 - 59세": "502",
     "60세 이상": "60",
     "60 - 64세": "601",
     "65세 이상": "602",
     "15 - 64세": "63",
     "15 - 24세": "70",
     "15 - 29세": "75"
}

```

```
prd_se = 'M'
load_gubun = '2'

query = query.format(key, itm_id['인구'], sex_id['계'], age_id['15세 이상 전체'], prd_se, load_gubun, org_id, tbl_id)
query
```

- 가장 최근 시점을 구했으니 시계열 조회를 한다.
- 통계청 OPEN API 사이트에서 시계열 조회 query 예제 복사

```
# 0:apiKey, 1:itmId, 2:objL1, 3:objL2, 4:prdSe, 5:startPrdDe, 6:endPrdDe, 7:loadGubun, 8:orgId, 9:tblId
query = "https://kosis.kr/openapi/Param/statisticsParameterData.do?method=getList&apiKey={0}&itmId={1}+&objL1={2}+&objL2={3}&objL3=&objL4=&objL5=&objL6=&objL7=&objL8=&format=json&jsonVD=Y&prdSe={4}&startPrdDe={5}&endPrdDe={6}&loadGubun={7}&orgId={8}&tblId={9}"

start_prd_de = '199906'
load_gubun = '1'
org_id
tbl_id

query.format(key, itm_id['인구'], sex_id['계'], age_id['15세 이상 전체'], prd_se, start_prd_de, end_prd_de, load_gubun, org_id, tbl_id )
```

```
# 자료 저장을 위한 DataFrame
df = pd.DataFrame()
```

- itm_id(항목)과 obj_lev2(연령구분)을 반복문으로 바꿔가며 조회하고, df에 모두 합친다.

```
for item in itm_id.values():
  for sex in sex_id.values():
    for obj in age_id.values():    
      query_temp = query.format(key, item, sex, obj, prd_se, start_prd_de, end_prd_de, load_gubun, org_id, tbl_id )        
      r = requests.get(query_temp)

      data = r.json()
      df_temp = pd.DataFrame(data)
      df = df.append(df_temp, ignore_index=True)
      print(item, sex, obj, df.shape)
      time.sleep(random.uniform(2,8))
```

- 합쳐진 결과를 csv 파일로 저장한다.(euc-kr 또는 utf-8-sig 추천)

```
df.to_csv('경제활동인구조사(euc-kr).csv', encoding='euc-kr')
files.download('경제활동인구조사(euc-kr).csv')
df.to_csv('경제활동인구조사(utf-8).csv', encoding='utf-8')
files.download('경제활동인구조사(utf-8).csv')
df.to_csv('경제활동인구조사(utf-8-sig).csv', encoding='utf-8-sig')
files.download('경제활동인구조사(utf-8-sig).csv')
```
